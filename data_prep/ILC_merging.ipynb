```python
import os
import scanpy as sc
from scipy import sparse
import numpy as np
import pandas as pd
import scipy.io as sio 
```

```python
pd.read_csv('D:/MyData/scEDGE_Generation/data/GSE70580/GSM1810510_T74_P1_A9_ILC1_expression.txt.gz',header=0, delimiter='\t')
```

```python
folder_path = 'D:/MyData/scEDGE_Generation/data/GSE70580'
```

```python
count_matrix = pd.DataFrame()  
```

```python
for file_name in os.listdir(folder_path):  
    if file_name.endswith('.txt.gz'):  # 确保只处理txt文件  
        file_path = os.path.join(folder_path, file_name)  
        # 读取文件  
        data = pd.read_csv(file_path, sep='\t', header=0)  
        # 提取第一列（Gene symbol）和第四列（Reads）  
        subset = data[['#Gene symbol', 'Reads']]  
        
        cell_name=file_name.split('.')[0]
        # 将当前文件的子集添加到count矩阵中  
        count_matrix[cell_name] = pd.Series(subset['Reads'].values, index=subset['#Gene symbol']).fillna(0)
```

```python
count_matrix
```

```python
cell_types = [col.split('_')[-2] for col in count_matrix.columns]
```

```python
cell_types
```

```python
cell_types_df = pd.DataFrame({  
    'cell_type': cell_types  
}) 
```

```python
counts=count_matrix.T
```

```python
cell_types_df.index=counts.index
```
